## Plotting curves

Let's start by plotting the curve $y = x^2$ from $x=0$ to $100$.

```python
import matplotlib.pyplot as plt
X = range(101)
Y = [value ** 2 for value in X]
plt.plot(X, Y)
plt.show()
```

## Plotting curves using NumPy

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2 * np.pi, num=100)
Y1 = np.sin(X)
Y2 = np.cos(X)
plt.plot(X, Y1)
plt.plot(X, Y2)
plt.show()
```

## Example

Plot the standard normal distribution ("bell curve") and its numerical derivative.


```python
import matplotlib.pyplot as plt
import numpy as np
  
X = np.linspace(-4, 4, 101)
Y = np.exp(-X**2 / 2) / (2 * np.sqrt(np.pi))
plt.plot(X, Y)

delta_X = np.diff(X)
delta_Y = np.diff(Y)
plt.plot(X[1:] - delta_X/2, delta_Y / delta_X)

plt.title('The normal distribution and its derivative')
plt.show()
```
